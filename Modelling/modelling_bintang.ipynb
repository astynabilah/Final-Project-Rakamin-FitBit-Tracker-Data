# Machine Learning

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np
import time
from sklearn.model_selection import cross_validate
```

# Function for model evaluation

```python
def evaluation_regressor(model, X_train, y_train, X_test, y_test):
    # Start timing the training process
    start_time_train = time.time()
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Predictions for both train and test data
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    # End timing
    end_time_train = time.time()
    train_time = end_time_train - start_time_train
    
    # Calculate evaluation metrics
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)
    mape_train = mean_absolute_percentage_error(y_train, y_pred_train)
    mape_test = mean_absolute_percentage_error(y_test, y_pred_test)
    
    # Get the model name in lowercase to use as a suffix
    model_name = model.__class__.__name__.lower()
    
    # Store metrics in a dictionary with model-specific keys
    metrics = {
        f'train_time_{model_name}': train_time,
        f'rmse_train_{model_name}': rmse_train,
        f'rmse_test_{model_name}': rmse_test,
        f'r2_train_{model_name}': r2_train,
        f'r2_test_{model_name}': r2_test,
        f'mape_train_{model_name}': mape_train,
        f'mape_test_{model_name}': mape_test
    }
    
    return metrics
```

## Linear Regression

```python
model = LinearRegression()
evaluation_regressor(model, X_train, y_train, X_test, y_test)

```

## Random Forest Regressor

```python
model = RandomForestRegressor(random_state=42)
evaluation_regressor(model, X_train, y_train, X_test, y_test)
```

## Decision Tree Regressor

```python
model = DecisionTreeRegressor(random_state=42)
evaluation_regressor(model, X_train, y_train, X_test, y_test)
```

## Support Vectore Machine Regressor

```python
model = SVR()
evaluation_regressor(model, X_train, y_train, X_test, y_test)
```

## Gradient Boosting

```python
model = GradientBoostingRegressor(random_state=42)
evaluation_regressor(model, X_train, y_train, X_test, y_test)
```

## XGBoost

```python
model = XGBRegressor(random_state=42)
evaluation_regressor(model, X_train, y_train, X_test, y_test)
```
